```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("default")

transactions = pd.read_excel("/content/transactions.xlsx")
products = pd.read_excel("/content/products.xlsx")


transactions.isna().sum()
products.isna().sum()

products["PRODUCT_ID"].nunique()

transactions["NET_SALES"] = (
    transactions["SALES_VALUE"]
    + transactions["RETAIL_DISC"]
    - transactions["COUPON_DISC"]
)

df = transactions.merge(
    products,
    on="PRODUCT_ID",
    how="left"
)

df.sort_values(by="NET_SALES", ascending=False).head(10)
df.sort_values(by="COUPON_MATCH_DISC", ascending=False).head(10)
```

```python
# Calculate purchase frequency for each product
product_purchase_counts = df['PRODUCT_ID'].value_counts().reset_index()
product_purchase_counts.columns = ['PRODUCT_ID', 'purchase_count']

# Merge with product details to get names and brands
most_frequent_products = product_purchase_counts.merge(
    products[['PRODUCT_ID', 'BRAND', 'COMMODITY_DESC']],
    on='PRODUCT_ID',
    how='left'
)

# Display the top 10 most frequently purchased products
display(most_frequent_products.head(10))

dept_sales = (
    df.groupby("DEPARTMENT")["NET_SALES"]
    .sum()
    .sort_values()
)

plt.figure()
dept_sales.plot(kind="barh")
plt.title("Total Net Sales by Department")
plt.xlabel("Net Sales")
plt.ylabel("Department")
plt.tight_layout()
plt.show()
df
```

```python
df['is_discounted'] = (
    (df['RETAIL_DISC'] < 0) |
    ((df['COUPON_DISC'] > 0) | (df['COUPON_MATCH_DISC'] > 0))
)
df['is_discounted'].value_counts()
```

```python
discount_comparison = df.groupby('is_discounted').agg(
    avg_net_sales=('NET_SALES', 'mean'),
    total_quantity=('QUANTITY', 'sum')
).reset_index()

display(discount_comparison)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(discount_comparison['is_discounted'].astype(str), discount_comparison['avg_net_sales'], color=['skyblue', 'lightcoral'])
plt.title('Average Net Sales by Discount Status')
plt.xlabel('Discount Status')
plt.ylabel('Average Net Sales')
plt.xticks(ticks=[0, 1], labels=['Non-Discounted', 'Discounted'])
plt.grid(axis='y', linestyle='--')
plt.show()

plt.figure(figsize=(10, 5))
plt.bar(discount_comparison['is_discounted'].astype(str), discount_comparison['total_quantity'], color=['lightgreen', 'orange'])
plt.title('Total Quantity Sold by Discount Status')
plt.xlabel('Discount Status')
plt.ylabel('Total Quantity Sold')
plt.xticks(ticks=[0, 1], labels=['Non-Discounted', 'Discounted'])
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
product_performance = df.groupby(['PRODUCT_ID', 'BRAND', 'COMMODITY_DESC']).agg(
    total_net_sales=('NET_SALES', 'sum'),
    total_quantity=('QUANTITY', 'sum')
).reset_index()
```

```python
top_10_revenue = product_performance.sort_values(by='total_net_sales', ascending=False).head(10)
display(top_10_revenue)
```
